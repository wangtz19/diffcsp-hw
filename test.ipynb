```python
import nest_asyncio
nest_asyncio.apply()

import asyncio
from playwright.async_api import async_playwright

async def main():
    async with async_playwright() as p:
        browser = await p.chromium.launch()
        print(f"Chromium version: {browser.version}")
        await browser.close()

        browser = await p.firefox.launch()
        print(f"Firefox version: {browser.version}")
        await browser.close()

        browser = await p.webkit.launch()
        print(f"Webkit version: {browser.version}")
        await browser.close()

asyncio.run(main())
```

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio
from playwright.async_api import async_playwright

async def main():
    async with async_playwright() as p:
        # browser = await p.chromium.launch() # Chromium
        # browser = await p.chromium.launch(channel="chrome") # Chrome
        # browser = await p.chromium.launch(channel="msedge") # Edge (Chromium)
        browser = await p.firefox.launch() # Firefox
        # browser = await p.webkit.launch() # webkit is Safari
        page = await browser.new_page()

        # Listen for all requests and print their headers
        page.on("request", lambda request: print("Request headers:", request.headers))

        # Listen for all responses and print their headers
        page.on("response", lambda response: print("Response headers:", response.headers))

        await page.goto("http://baidu.com")
        print(await page.title())
        await browser.close()

asyncio.run(main())
```
